```python
import ollama

def answer_ai(msg,model='deepseek-r1:1.5b'):
    response = ollama.chat(model=model, messages=[
      {
        'role': 'user',
        'content': msg,
      },
    ])
    return response['message']['content']
```

```python
model = 'deepseek-r1:32b'
```

```python
msg = "why is the sky blue"

print(answer_ai(msg,model))
```
